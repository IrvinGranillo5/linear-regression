# üñãÔ∏èIntroducci√≥n
La media aritm√©tica es una medida de tendencia central que representa el valor <span style="color:deepskyblue">promedio</span>
 de un conjunto de datos. Se utiliza ampliamente en estad√≠stica, an√°lisis de datos y ciencias aplicadas para resumir informaci√≥n num√©rica.

## Explicaci√≥n Detallada de la Media.
La la media se presenta como la medida de tendencia central m√°s cn.". Se define matem√°ticamentera un conjunto de datoas $x_1, x_2, \dots, x_n$

Representado con la siguiente formula:

o$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i \quad $_



Donde:

$\bar{x}$ es la media (denotada com√∫nmente con una barra sobre la variable).


$n$ es el n√∫mero total de observaciones o datos en el conjunto.


$\sum_{i=1}^{n} x_i$ representa la suma de todos los valores individuales.

En resumen:

para obtener la media o promedio se debera sumar todo el conjunto de los datoa a promediar, dividirlo entre los n elementos que este contenga.4 + 6 = 12\).




## üßæ Resumen del conjunto de datos

En este ejemplo trabajaremos con un conjunto de datos **simulado y almacenado como un arreglo multidimensional (`numpy.ndarray`)**, ideal para an√°lisis estad√≠stico, visualizaci√≥n y documentaci√≥n t√©cnica.

El objetivo es generar datos veros√≠miles de temperaturas diarias en cuatro ciudades distintas, durante un periodo prolongado de 1,000 d√≠as. Esta simulaci√≥n permite practicar t√©cnicas de an√°lisis sin depender de fuentes externas, manteniendo control total sobre la estructura y variabilidad de los datos.

### üîß Caracter√≠sticas del conjunto:

- **Generaci√≥n sint√©tica**: Se utiliza una funci√≥n seno para simular ciclos estacionales y ruido gaussiano para variaciones diarias.
- **Reproducibilidad**: Se fija una semilla (`np.random.seed`) para obtener resultados consistentes.
- **Formato estructurado**: Los datos se almacenan en un arreglo `numpy.ndarray` de forma `(1000, 4)`, donde cada columna representa una ciudad.
- **Variables simuladas**:
  - Ciudad Costera: clima templado con variaciones suaves.
  - Ciudad Monta√±osa: clima fr√≠o con menor amplitud t√©rmica.
  - Ciudad Continental: clima extremo con alta variabilidad.
  - Ciudad Tropical: clima c√°lido y estable con ruido moderado.



.



```python
import numpy as np
import pandas as pd

    #Fijamos la semilla para reproducibilidad
np.random.seed(616)

# N√∫mero de d√≠as simulados
dias = 1000

# simulaci√≥n de variaci√≥n estacional con ruido
# Usamos una funci√≥n seno para simular ciclos anuales y agregamos ruido normal

#Ciudad costera: clima templado con variaciones suaves
costera = 20 + 8 * np.sin(np.linspace(0, 4 * np.pi, dias)) + np.random.normal(0, 2, dias)

#ciudad monta√±osa: clima fr√≠o con menor amplitud t√©rmica
monta√±osa = 10 + 4 * np.sin(np.linspace(0, 4 * np.pi, dias)) + np.random.normal(0, 1.5, dias)

#Ciudad continental: clima extremo con mayor variaci√≥n estacional
continental = 15 + 12 * np.sin(np.linspace(0, 4 * np.pi, dias)) + np.random.normal(0, 3, dias)

# Ciudad tropical: clima c√°lido y estable con ruido moderado
tropical = 25 + 3 * np.sin(np.linspace(0, 4 * np.pi, dias)) + np.random.normal(0, 2.5, dias)

# Guardamos los datos en un arreglo numpy.ndarray de forma (1000, 4)
temperaturas = np.round(np.column_stack([costera, monta√±osa, continental, tropical]), 1)

# Creamos un DataFrame para visualizaci√≥n r√°pida, no es necesario para el ejercicio pero nos ayudaara a visualizar nuestros datos 
df = pd.DataFrame(temperaturas, columns=[
    "Costera (¬∞C)", "Monta√±osa (¬∞C)", "Continental (¬∞C)", "Tropical (¬∞C)"
])
df["D√≠a"] = np.arange(1, dias + 1)

# Mostramos las primeras filas
#df

```

## Problema 2
Calcula el resto de medias utilizando una funcion Llamada `media(x)` y guarda los valores en `X2`,`X3`,`X4` respectivamente, recuerda esto se debera hacer sin pcupar <span style="color:red">.sum() o .mean() </span>

```python
### BEGIN SOLUTION
def media(datos):
    sum=0
    for s in datos:
        sum+=s
        
    return sum/ len(datos)
X2 = media(monta√±osa)
X3 = media(tropical)
X4 =media(continental)
### END SOLUTION

print(f"La media de la zona monta√±osa es : {X2}")
print(f"La media de la zona monta√±osa es : {X3}")
print(f"La media de la zona monta√±osa es : {X4}")
```
