## Definici贸n

La **desviaci贸n est谩ndar** es una medida de dispersi贸n que indica cu谩nto se alejan, en promedio, los datos respecto a la media aritm茅tica. Es especialmente 煤til para evaluar la variabilidad de un conjunto de datos: cuanto mayor sea la desviaci贸n est谩ndar, m谩s dispersos est谩n los valores.

Se representa com煤nmente con la letra griega sigma (****) para la poblaci贸n, o con **s** para la muestra.

Por ejemplo, si tenemos el siguiente conjunto de valores $X = \{3, 5, 1, 2\}$, primero calculamos la media:

$$
\bar{X} = \frac{3 + 5 + 1 + 2}{4} = \frac{11}{4} = 2.75
$$

Luego, calculamos la desviaci贸n est谩ndar poblacional:

$$
\sigma = \sqrt{ \frac{(3 - 2.75)^2 + (5 - 2.75)^2 + (1 - 2.75)^2 + (2 - 2.75)^2}{4} } = \sqrt{ \frac{0.0625 + 5.0625 + 3.0625 + 0.5625}{4} } = \sqrt{2.6875} \approx 1.64
$$

## **Definici贸n.**

Dado el conjunto de datos $X = \{x_1, \dots, x_N\}$, la desviaci贸n est谩ndar se define como sigue:



---

###  Desviaci贸n est谩ndar para poblaci贸n completa


$$\sigma = \sqrt{ \frac{\sum (x_i - \mu)^2}{N} }$$


---

###  Desviaci贸n est谩ndar para muestra

<div style="color:#003366; font-weight:bold; font-size:16px">
$$s = \sqrt{ \frac{\sum (x_i - \bar{x})^2}{n - 1} }$$
</div>

---

### Donde:

- $x_i$ = cada valor individual del conjunto de datos  
- $\mu$ = media poblacional  
- $\bar{x}$ = media muestral  
- $N$ = tama帽o total de la poblaci贸n  
- $n$ = tama帽o de la muestra*
